```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0,10.0)

#Reading Data
data = pd.read_csv('Salary_Data.csv')
print(data.shape)
data.head()
```

```python
X = data['YearsExperience'].values
Y = data['Salary'].values

mean_x = np.mean(X)
mean_y =np.mean(Y)

m = len(X)

numer = 0
denom = 0
for i in range(m):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x)**2
    
b1 = numer/denom
b0 = mean_y - (b1*mean_x)
    
    
print(b1,b0)
```

```python
max_x = np.max(X)
min_x = np.min(X) 

x = np.linspace(min_x, max_x, 10)
y = b0 + b1*x

#Plotting

plt.plot(x,y, color='#58b970' , label='Regression Line')
plt.scatter(X,Y, c='#ef5423', label = 'Scatter Plot')

plt.xlabel('YersExp')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
ss_t =0
ss_r =0
for i in range(m):
    y_pred = b0 + b1*X[i]
    ss_t +=(Y[i] - mean_y)**2
    ss_r +=(Y[i] - y_pred)**2

r2 = 1-(ss_r/ss_t)
print(r2)    
```
